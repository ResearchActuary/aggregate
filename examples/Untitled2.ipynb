```python
def dq(x):
    return quote(quote(x, safe=''), safe='')

def duq(x):
    return unquote(unquote(x))
```

```python
def pget(type_, iri_id):
    dd = {'individuals': 'https://www.ebi.ac.uk/ols/api/ontologies/probonto/individuals',
          'properties':  'https://www.ebi.ac.uk/ols/api/ontologies/probonto/properties',
          'self':        'https://www.ebi.ac.uk/ols/api/ontologies/probonto',
          'terms':       'https://www.ebi.ac.uk/ols/api/ontologies/probonto/terms'}
    iri = f'http://www.probonto.org/ontology#{iri_id}'
    url =  f'{dd[type_]}/{dq(iri)}'
    r = requests.get(url)
    return r.json()
```

```python
def search(term):
    r = requests.get(f'https://www.ebi.ac.uk/ols/api/search?q={term}&ontology=probonto')
    d = r.json()
    return d['response']['docs']
```

```python
search('lognormal')
```
